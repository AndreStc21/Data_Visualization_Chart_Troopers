```python
import pandas as pd
import numpy as np

df_emission_type = pd.read_csv('co2-fossil-plus-land-use.csv')
df_emission_type = df_emission_type[~((df_emission_type['Code'].isna())|(df_emission_type['Code']=='OWID_WRL'))]
df_emission_type = df_emission_type.rename(columns={
    'Annual CO₂ emissions including land-use change': 'Annual CO2 emissions including land-use change',
    'Annual CO₂ emissions from land-use change': 'Annual CO2 emissions from land-use change',
    'Annual CO₂ emissions': 'Emissions'
    })
df_emission_type = df_emission_type[df_emission_type['Year']==max(df_emission_type['Year'])]
df_emission_type[df_emission_type["Emissions"].isna()] = 0
for c in ['Emissions']:
    df_emission_type[c] /= 1_000_000
    df_emission_type[c] = df_emission_type[c].round(0)
    df_emission_type[c] = df_emission_type[c].astype(int)
df_emission_type = df_emission_type[['Entity', 'Code', 'Emissions']]
df_emission_type
```

```python
df_emission_type.to_csv('co2-total-country-emissions.csv', index=False)
```

```python
df_emission_per_capita = pd.read_csv('co2-emissions.csv')
df_emission_per_capita = df_emission_per_capita[~((df_emission_per_capita['Code'].isna())|(df_emission_per_capita['Code']=='OWID_WRL'))]
df_emission_per_capita = df_emission_per_capita.rename(columns={
    'Annual CO₂ emissions (per capita)': 'Emissions'
    })
df_emission_per_capita = df_emission_per_capita[df_emission_per_capita['Year']==max(df_emission_per_capita['Year'])]
df_emission_per_capita[df_emission_per_capita["Emissions"].isna()] = 0
df_emission_per_capita = df_emission_per_capita[['Entity', 'Code', 'Emissions']]
df_emission_per_capita
```
