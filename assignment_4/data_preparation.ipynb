```python
import pandas as pd
import numpy as np

df_avg_monthly_temperatures_per_year = pd.read_csv('avg.csv')
df_max_monthly_temperatures_per_year = pd.read_csv('max.csv')
df_min_monthly_temperatures_per_year = pd.read_csv('min.csv')
df_avg_monthly_temperatures_per_year
```

```python
def split_date(df):
    df['Date'] = df['Date'].astype(str)
    df['year'] = df['Date'].str[:4] # First 4 characters
    df['month'] = df['Date'].str[4:] # Last 2 characters

    # Optionally, convert 'year' and 'month' columns back to integers
    df['year'] = df['year'].astype(int)
    df['month'] = df['month'].astype(int)
    return df
```

```python
def prepare_table(df):
    # Create 'year' and 'month' columns
    df_no_anomaly = df.drop(columns = "Anomaly")
    df_split_date = split_date(df_no_anomaly)
    return df_split_date
    
```

```python
df_avg_monthly_temperatures_per_year = prepare_table(df_avg_monthly_temperatures_per_year)
df_max_monthly_temperatures_per_year = prepare_table(df_max_monthly_temperatures_per_year)
df_min_monthly_temperatures_per_year = prepare_table(df_min_monthly_temperatures_per_year)
df_avg_monthly_temperatures_per_year
```

```python
df_avg_monthly_temperatures_per_year.to_csv('monthly_avg_per_year.csv', index=False)
df_max_monthly_temperatures_per_year.to_csv('monthly_max_per_year.csv', index=False)
df_min_monthly_temperatures_per_year.to_csv('monthly_min_per_year.csv', index=False)
```
